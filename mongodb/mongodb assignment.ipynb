```python
# Import libraries
import json
import pymongo
from pymongo import MongoClient

from datetime import datetime
```

```python
# Connect to Mongo DB
myclient = pymongo.MongoClient("mongodb://localhost:27017/")
new_db   = myclient["mydatabase"]

print(myclient.list_database_names())
```

# Click Logs

```python
# new_db['clicks'].drop()
# new_db['coffee'].drop()
```

```python
file_loc = 'data/click_log.json'

clicks_docs = []
with open(file_loc, 'r', encoding="utf8") as file:
    line = file.readline()
    while line:
        clicks_docs.append(json.loads(line))
        line = file.readline()

len(clicks_docs)
```

```python
clicks_docs[25]
```

```python
# Create click collection in mydatabase
clicks_col = new_db['clicks']
```

```python
# Insert many documents into clicks collection
clicks_01 = clicks_col.insert_many(clicks_docs)

#print list of the _id values of the inserted documents:
print(clicks_01.inserted_ids)
```

```python
# Count how many documents are in collection
clicks_col.find().count()
```

```python
# Show the first 10
result = clicks_col.find().limit(10)

#print the result:
count = 1
for x in result:
    print(str(count) + ': ', x, '\n')
    count +=1
```

```python
# Find documents where city is San Francisco
query = {'cy':'San Francisco'}
newdoc = clicks_col.find(query)

#print the result:
for x in newdoc:
    print(x['cy'])


# Print document count
print(clicks_col.find(query).count())
```

```python
# Find distinct browsers
browsers = clicks_col.distinct('a')

print(len(browsers))
```

```python
# Documents that are Mozilla or Opera
query = {'$or': [{'a': {'$regex':'^Mozilla'}}, {'a':{'$regex': '^Opera'}}]}

result = clicks_col.find(query)
for x in result:
    print(x['a'], '\n')
```

```python
# Find type of timestamp
type(clicks_col.find_one({'t': {'$exists': True}})['t'])
```

```python
# Convert to datetime
query = {'t':{'$exists': True}}
result = clicks_col.find(query).sort("t")

# Create date and hour key values
for x in result:
    dt     = datetime.fromtimestamp(x['t'])
    #dt_str = dt.strftime("%Y/%m/%d, %H:%M:%S")
    #hour   = int(dt.strftime("%H"))

    query  = { "_id": x["_id"] }
    newvalues = { "$set": { "date": dt } }
    clicks_col.update_one(query, newvalues)
    
# Display t, date and hour
query = {'date':{'$exists': True}}
result2 = clicks_col.find(query).sort("t").limit(10)

for x in result2:    
    print(x['t'], x['date'])
```

```python
# Sort documents by timestamp
query = {'t':{'$exists': True}}
result = clicks_col.find(query).sort("t").limit(1)

for x in result:
    first_ts = x['t']
#     print(x, '\n')

# Get the next hour timestamp
last_ts = first_ts+ 360
query = {'t': {'$lte': last_ts}}
result = clicks_col.find(query).sort("t").limit(10)

# Print the results
for x in result:
    print(x, '\n')
```

```python
# Count each url and then sort
# Count browsers
result = clicks_col.aggregate([
    # Group the documents and "count" via $sum on the values
    { '$group': { '_id': "$u", 'count': { '$sum': 1} } },
    { '$sort': { 'count': -1 }}
])

control = 0
for x in result:
    if(control == 0):
        print(x['_id'], ':', x['count'])
        control+=1
    else:
        break
```
